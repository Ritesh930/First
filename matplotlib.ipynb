```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
```

```python
arr=np.random.randint(20000,40000,100)
arr
```

```python
#lineplot
arr.sort()
plt.plot(arr)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,10,9,30,25]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,10,9,30,25]
plt.plot(x,y,marker='o') #marker='o' adds circle markers at each data point
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,10,9,30,25]
plt.plot(x,y,marker='o',ms=10,
         mfc="yellow",mec="green",color="black",lw=3,ls="--",alpha=0.5,label="lineplot") #marker='o' adds circle markers at each data point,marker size is set to 10
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
#scatterplot----work on nmerical data
plt.scatter(x,y,marker="s")  #marker="s" adds square markers at each data point
plt.title("Scatter Plot")   # title of the plot
plt.show()
```

```python
x=["mp","up","rj","tn","mh"]
y=[10,20,30,40,50]
colours=["red","green","blue","yellow","orange"]
hatch_patterns=["/","\\","|","-","*"]
plt.bar(x,y,color=colours,hatch=hatch_patterns,
       label = x,width=0.3,edgecolor="black",alpha=0.5 )  #bar plot
#plt.barh(x,y) #horizontal bar plot
plt.legend()
plt.show()
```

```python
#  Histogram----nuumerical data, distribution in intervals(bins)
#  piechart----categorical data
arr=np.random.randint(20000,40000,100)
plt.hist(arr,bins=20,edgecolor="black",alpha=0.5)  #histogram with 10 bins
plt.show()

```

```python
x=["mp","up","rj","tn","mh"]
y=[10,20,30,40,50]
explodes=(0.1,0,0,0,0)  #explode the first slice
plt.pie(y,labels=x,autopct="%1.2f%%",explode=explodes,)  #pie chart with percentage labels,explode mp part
plt.legend()
plt.savefig("pie_chart.png")  #save the pie chart as an image
plt.show()
```

```python
## subplot
x=np.linspace(0,10,100)
plt.Figure(figsize=(10,6))  #set figure size
plt.subplot(2,2,1)  #2 rows, 2 columns, first subplot
plt.plot(x,np.sin(x))
    
plt.subplot(2,2,2)  #2 rows, 2 columns, second subplot
plt.plot(x,np.cos(x))                                                       
plt.subplot(2,2,3)  #2 rows, 2 columns, third subplot
plt.plot(x,np.tan(x))

plt.subplot(2,2,4)  #2 rows, 2 columns, fourth
plt.plot(x,np.log(x))
plt.show()
import warnings
warnings.filterwarnings("ignore")  #ignore warnings
```
